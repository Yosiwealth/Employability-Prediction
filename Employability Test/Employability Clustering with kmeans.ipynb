```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('employability2.csv')
neuro = pd.read_csv('neuroticism.csv')
```

```python
neuro.head(3)
```

```python
X = df[['OPENNESS','CONSCIENTIOUS','EXTROVERSION','AGREEABLE','NEUROTICISM']]
y = df['employ']
```

```python
nX = neuro[['N1','N2','N3','N4','N5','N6','N7','N8','N9','N10']]
ny = neuro['NEUROTICISM']
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2)
kmNeuro = KMeans(n_clusters=2)

df['Emp_Clus'] = km.fit_predict(X)
neuro['Neu_Clus'] = kmNeuro.fit_predict(nX)
```
